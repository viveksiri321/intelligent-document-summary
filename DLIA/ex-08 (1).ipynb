```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img 
img = load_img('WhatsApp Image 2025-07-29 at 14.16.42_d2197d68.jpg')  
from numpy import expand_dims 
from keras.preprocessing.image import load_img 
from keras.preprocessing.image import img_to_array 
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
import matplotlib.pyplot  as plt
```

```python
# load the image 
plt.figure(figsize=(45,30)) 
 
# convert to numpy array 
data = img_to_array(img) 
```

```python
 
# expand dimension to one sample 
samples = expand_dims(data, 0) 
```

```python
# create image data augmentation generator 
datagen = ImageDataGenerator(featurewise_center=True,rotation_range=(0-30),width_shift_range=0.2, 
                             height_shift_range=0.2,brightness_range=[0.5,1.5], 
                             shear_range=0.2, zoom_range=0.2,channel_shift_range=0.2, 
                             horizontal_flip=True, vertical_flip=True,fill_mode='nearest') 
```

```python
# prepare iterator 
it = datagen.flow(samples, batch_size=1) 
 
# generate samples and plot 
for i in range(6): 
    # define subplot 
    plt.subplot(330 + 1 + i) 
    # generate batch of images 
batch = next(it) # Use next(it) instead of it.next() 
# convert to unsigned integers for viewing 
image = batch[0].astype('uint8') 
# plot raw pixel data 
plt.imshow(image) 
# show the figure 
plt.show()
```

```python
# create image data augmentation generator 
datagen = ImageDataGenerator(featurewise_center=True,rotation_range=(0-30),width_shift_range=0.2, 
                             height_shift_range=0.2,brightness_range=[0.5,1.5], 
                             shear_range=0.2, zoom_range=0.2,channel_shift_range=0.2, 
                             horizontal_flip=True, vertical_flip=True,fill_mode='nearest') 
# prepare iterator 
it = datagen.flow(samples, batch_size=1) 
# Generate and plot augmented images
plt.figure(figsize=(10, 6))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    batch = next(it)
    augmented_image = batch[0].astype('uint8')
    plt.imshow(augmented_image)
    plt.axis('off')

plt.tight_layout()
plt.show()
 
```
