```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from tensorflow.keras.datasets import cifar10 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense 
from tensorflow.keras.utils import to_categorical 
```

```python
def create_model(): 
    model = Sequential([ 
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)), 
        MaxPooling2D((2, 2)), 
        Flatten(), 
        Dense(128, activation='relu'), 
        Dense(10, activation='softmax') 
    ]) 
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 
    return model 
```

```python
# Load CIFAR-10 dataset 
(x_train, y_train), (x_test, y_test) = cifar10.load_data() 
 
# Preprocess data 
x_train = x_train.astype('float32') / 255  
x_test = x_test.astype('float32') / 255    
y_train = to_categorical(y_train, num_classes=10) 
y_test = to_categorical(y_test, num_classes=10) 
```

```python
# Define split ratios 
split_ratios = [0.7, 0.8, 0.9] 
 
for split_ratio in split_ratios: 
    # Split data into train, validation, and test sets 
    # Use x_train and y_train instead of x and y 
    x_train_split, x_temp, y_train_split, y_temp = train_test_split(x_train, y_train, test_size=1 - split_ratio, random_state=42) 
    x_val, x_test_split, y_val, y_test_split = train_test_split(x_temp, y_temp, test_size=0.2, random_state=42) 
 
    model = create_model() 
    history = model.fit(x_train_split, y_train_split, epochs=10,validation_data=(x_val, y_val)) 
     # Evaluate on the test set 
    test_loss, test_acc = model.evaluate(x_test_split, y_test_split) 
    print(f"Split ratio: {split_ratio}, Test accuracy: {test_acc:.4f}")
     # Plot training and validation accuracy 
    plt.plot(history.history['accuracy'], label='train') 
    plt.plot(history.history['val_accuracy'], label='val') 
    plt.title(f'Accuracy for split ratio {split_ratio}') 
    plt.xlabel('Epoch') 
    plt.ylabel('Accuracy') 
    plt.legend() 
    plt.show()
```
